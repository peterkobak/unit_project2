## Intermediate level

#### Count how many movies have each of the content ratings.

```python
# Answer:
movies.loc[:,'content_rating'].value_counts()
```

#### Use a visualization to display that same data, including a title and x and y labels.

```python
# Answer:
movies.loc[:, 'content_rating'].value_counts().plot(kind='bar', figsize=(12,12));
plt.xlabel('Content Rating')
plt.ylabel('Number of Movies')
plt.title('IMDB Movies by Content Rating')
```

#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
# Answer:
movies.loc[:,'content_rating'].replace('NOT RATED', 'UNRATED', inplace=True)
movies.loc[:,'content_rating'].replace('APPROVED', 'UNRATED', inplace=True)
movies.loc[:,'content_rating'].replace('PASSED', 'UNRATED', inplace=True)
movies.loc[:,'content_rating'].replace('GP', 'UNRATED', inplace=True)
movies
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
# Answer:
movies.loc[:, 'content_rating'].replace('X', 'NC-17', inplace=True)
movies.loc[:, 'content_rating'].replace('TV-MA', 'NC-17', inplace=True)
movies.loc[:, 'content_rating'].value_counts()
```

#### Count the number of missing values in each column.

```python
# Answer:
movies.isnull().sum()
```

#### If there are missing values: examine them, then fill them in with "reasonable" values.

```python
# Answer:

# Check which are null and examine them
which_are_null = movies.loc[:,'content_rating'].isnull()
movies.loc[which_are_null,:]

# Replace those values with 'UNRATED'
movies.loc[which_are_null,'content_rating'] = 'UNRATED'
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Answer:
movies['over_2_hours'] = (movies.loc[:,'duration'] >= 120)
movies.groupby('over_2_hours')['star_rating'].mean()
```

#### Use a visualization to detect whether there is a relationship between duration and star rating.

#### Calculate the average star rating for each genre, but only include genres with at least 10 movies


```python
#Answer

# Return a list of genres that contain over 10 movies
dataframe_of_genres_value_counts = ((movies.loc[:,'genre'].value_counts() > 10).reset_index())
genres_with_over_10_movies = list(dataframe_of_genres_value_counts.loc[
    (dataframe_of_genres_value_counts.loc[:,'genre'] == True),'index'
])

# Create an index of movies that are of a genre with over 10 movies
movies_in_genres_with_over_10_movies = movies.loc[:,'genre'].apply(lambda genre: genre in genres_with_over_10_movies)

# Calculate the average star rating of the genres, based on the previous index of movies that 
# are of a genre with over 10 movies
movies.loc[movies_in_genres_with_over_10_movies, :].groupby('genre')['star_rating'].mean()
```
